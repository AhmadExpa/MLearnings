### Linear Regression

```python
# Importing Libraries
import csv
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv")
df = df.dropna()
```

### Step-2 Splitting Dataset into Training Data and Testing Data

```python
x = df[['Year of Experience']]
y = df['Salary']
```

```python
df.columns
```

```python
# import library
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=0)
```

## Step -3 Fit Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
data_model = LinearRegression().fit(x_train, y_train)
data_model
```

```python
 import matplotlib.pyplot as plt
plt.scatter(x_train, y_train)
plt.plot(x_train, data_model.predict(x_train), color='red')
plt.xlabel("Experinece Of Year")
plt.ylabel("Salary")
plt.title('Train Data')
plt.show()#
```

## Step-5 Testing or Evaluating Your Data

```python
# Model Fitness
print("Traning Data Score = ",data_model.score(x_train, y_train))
print("Testing Data Score = ",data_model.score(x_test, y_test))
```
